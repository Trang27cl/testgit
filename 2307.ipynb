```python
import math
xa=float(input("Nhap hoanh do diem A: "))
ya=float(input("Nhap tung do diem A: "))
xb=float(input("Nhap hoanh do diem B: "))
yb=float(input("Nhap tung do diem B: "))
xc=float(input("Nhap hoanh do diem C: "))
yc=float(input("Nhap tung do diem C: "))
#phuong trinh duong thang AB co dang: ax+by+c=0
#Vecto phap tuyen duong thang AB:
aAB=-(yb-ya)
bAB=xb-xa
#duong thang di qua diem A: 
cAB= -aAB*xa-bAB*ya
d= aAB*xc+bAB*yc+cAB==0
print("Khong la tam giac: ",d)
#phuong trinh duong thang AC co dang: ax+by+c=0
#Vecto phap tuyen duong thang AC:
aAC=-(yc-ya)
bAC=xc-xa
#duong thang di qua diem A: 
cAC= -aAC*xa-bAC*ya
#phuong trinh duong thang BC co dang: ax+by+c=0
#Vecto phap tuyen duong thang BC:
aBC=-(yc-yb)
bBC=xc-xb
#duong thang di qua diem B: 
cBC= -aBC*xb-bBC*yb
#Goi D,E,F lan luot la trung diem AB,AC,BC
xd=(xa+xb)/2
yd=(ya+yb)/2
xe=(xa+xc)/2
ye=(ya+yc)/2
xf=(xb+xc)/2
yf=(yb+yc)/2
CD=math.sqrt((xd-xc)**2+(yd-yc)**2)
BE=math.sqrt((xe-xb)**2+(ye-yb)**2)
AF=math.sqrt((xf-xa)**2+(yf-ya)**2)
print("do dai cac trung tuyen CD, BE,AF lan luot la ",CD,BE,AF)
#Goi W la giao diem 3 duong trung tuyen
xw=(2*(xa+xf))/3
yw=(2*(ya+yf))/3
print("Toa do giao diem 3 duong trung tuyen la ","xw= "+str(xz),"yw= "+str(yz))
#Goi K,M,N lan luot la giao diem cac duong cao voi 3 canh AB,AC,BC
#phuong trinh duong thang CK co dang: ax+by+c=0
#Vecto phap tuyen duong thang CK:
aCK= xb-xa
bCK=yb-ya
#duong thang di qua diem C: 
cCK= -aCK*xc-bCK*yc
p=-aAB/aCK
yk=(-p*cCK-cAB)/(bAB+bCK*p)
xk=(-cCK-bCK*yk)/aCK
CK=math.sqrt((xk-xc)**2+(yk-yc)**2)
#phuong trinh duong thang BM co dang: ax+by+c=0
#Vecto phap tuyen duong thang BM:
aBM=xc-xa
bBM=yc-ya
#duong thang di qua diem B: 
cBM= -aBM*xb-bBM*yb
q=-aAC/aBM
ym=(-q*cBM-cAC)/(bAC+bBM*q)
xm=(-cBM-bBM*ym)/aBM
BM=math.sqrt((xm-xb)**2+(ym-yb)**2)
#phuong trinh duong thang AN co dang: ax+by+c=0
#Vecto phap tuyen duong thang AN:
aAN=xc-xb
bAN=yc-yb
#duong thang di qua diem A: 
cAN= -aAN*xa-bAN*ya
j=-aBC/aAN
yn=(-j*cAN-cBC)/(bBC+bAN*j)
xn=(-cAN-bAN*yn)/aAN
AN=math.sqrt((xn-xa)**2+(yn-ya)**2)
print("do dai cac duong cao CK, BM,AN lan luot la ",CK,BM,AN)
#Goi Z la giao diem 3 duong cao
r=-aBM/aAN
yz=(-r*cAN-cBM)/(bBM+bAN*r)
xz=(-cAN-bAN*yz)/aAN
print("Toa do giao diem 3 duong cao la ","xz= "+str(xz),"yz= "+str(yz))
import turtle
t=turtle.Turtle()
t.shape('turtle')
t.hideturtle()
maunetve=input("Nhap mau net ve: ")
t.color(maunetve)
t.penup()
t.goto(xa,ya)
t.pendown()
t.dot(3)
t.write("A",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xb,yb)
t.dot(3)
t.write("B",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xc,yc)
t.dot(3)
t.write("C",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xa,ya)
t.goto(xf,yf)
t.dot(3)
t.write("F",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xb,yb)
t.goto(xe,ye)
t.dot(3)
t.write("E",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xc,yc)
t.goto(xd,yd)
t.dot(3)
t.write("D",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xa,ya)
t.goto(xn,yn)
t.dot(3)
t.write("N",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xb,yb)
t.goto(xm,ym)
t.dot(3)
t.write("M",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xc,yc)
t.goto(xk,yk)
t.dot(3)
t.write("K",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xz,yz)
t.dot(3)
t.write("Z",move=False,align='left',font=('Arial', 8, 'normal'))
t.goto(xa,ya)
t.goto(xw,yw)
t.dot(3)
t.write("W",move=False,align='left',font=('Arial', 8, 'normal'))
t=turtle.done()
```
