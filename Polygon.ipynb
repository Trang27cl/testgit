```python
import turtle as t


class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y


def drawStart(x, y, text, ix, iy, color, icolor):
    t.speed(8)
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.pensize(2)
    t.pencolor('blue')
    t.fillcolor(color)
    t.begin_fill()
    for i in range(2):
        t.forward(100)
        t.left(90)
        t.forward(30)
        t.left(90)
    t.end_fill()
    t.penup()
    t.goto(x + ix, y + iy)
    t.pencolor(icolor)
    t.write(text, font=('Courier', 12, 'normal'))


def visualize(points, color='red', width=2):
    if len(points) == 0:
        print('please set points')
        return
    t.penup()
    t.goto(points[0].x, points[0].y)
    t.pendown()
    t.pencolor(color)
    t.pensize(width)
    for i in range(1, len(points)):
        t.goto(points[i].x, points[i].y)
    t.goto(points[0].x, points[0].y)


def estimate(points):
    if len(points) == 0:
        print('please set points')
        return
    isconvex = True
    clone_points = points.copy()
    clone_points.append(points[0])
    for i in range(len(clone_points) - 1):
        exp = 0
        for p in points:
            if (p.x == clone_points[i].x and p.y == clone_points[i].y) \
                    or (p.x == clone_points[i + 1].x and p.y == clone_points[i + 1].y):
                continue
            a = (clone_points[i + 1].y - clone_points[i].y) * (p.x - clone_points[i].x)
            b = (clone_points[i + 1].x - clone_points[i].x) * (p.y - clone_points[i].y)
            if exp == 0:
                exp = a - b
            else:
                if exp * (a - b) < 0:
                    isconvex = False
                    break
    t.penup()
    root_x = 0
    root_y = 0
    for p in points:
        root_x += p.x
        root_y += p.y
    root_x /= len(points)
    root_y /= len(points)
    t.goto(root_x-50, root_y)    
    t.pensize(2)
    if isconvex:
        t.pencolor('green')
        t.write('Đa giác lồi', font=('Courier', 12, 'normal'))
    else:
        t.pencolor('red')
        t.write('Đa giác lõm', font=('Courier', 12, 'normal'))


points = []
drawStart(-320, 250, 'Draw', 25, 8, 'red', 'white')

drawStart(-200, 250, 'Estimate', 13, 8, 'orange', 'white')

t.pensize(3)
t.speed(8)
t.hideturtle()
isChecking = False


def setpoint(i, j):
    global points
    if j >= 250:
        if (i >= -320) and (i <= -220):
            print('Set points')
            visualize(points, color='blue')
        if (i >= -200) and (i <= -100):
            print('Checking')
            estimate(points)
            points = []
        return
    print(f'x = {i}, y = {j}')
    t.penup()
    t.goto(i, j - 1.5)
    t.pencolor('red')
    t.pendown()
    t.speed(8)
    t.shape('circle')
    t.pensize(2)
    t.stamp()
    points.append(Point(i, j))


t.onscreenclick(setpoint)

t.done()
```
