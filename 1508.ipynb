```python
#BTVN2 12/08
def max_min_in_text(text):
    '''Hàm tìm even min and odd max của các số có trong chuỗi
    Parameters:
    - text(str): một chuỗi bất kỳ có chứa các chữ số cách nhau bởi dấu cách
    Returns:
    - max(int): Số chẵn lớn nhất
    - min(int): Số lẻ nhỏ nhất
    '''
    text=text.strip()
    words=text.split()
    _max = None
    _min = None
    for w in words:
        if w.isdigit():
            if int(w)%2==0:
                if _max==None:
                    _max = int(w)
                else:
                    if _max<int(w):
                        _max = int(w)
            else:
                if _min==None:
                    _min = int(w)
                else:
                    if _min>int(w):
                        _min = int(w)
    return _max,_min
x,y=max_min_in_text('abd 34 tr13 85 sh -3423 21 def')
print('Max = ', x if x!=None else 'Không có số chẵn')
print('Min = ',y)
```
