```python
from tkinter import *
top = Tk()
top.geometry("450x300")
def changeText():
    #lấy text nhập từ Entry
    text = myinput.get()
    text = text.upper()
    #Hiện thị text nhận được lên label
    user_label['text'] = text
    #Hiển thị ngay luôn trên entry
    myinput.delete(0,END)
    myinput.insert(0,text)
#photo = top.PhotoImage(file = 'error.jpg')
myfont = ('Helvetica','16')
user_label = Label(top,text = 'Python programming', bg = 'blue',fg = 'red', font=myfont)
user_label.place(x=40,y =100)
#ô nhập dữ liệu Entry - chỉ nhập 1 dòng
myinput = Entry (top, font = myfont, justify = RIGHT)
myinput.pack()
button = Button(top, text = 'Change text', command = changeText)
button.pack()
top.mainloop()
```

```python
import tkinter as tk
window = tk.Tk()
window.title('Tính tổng 2 số')
window.geometry('250x300')
def kq():
    text_so1= so1.get()
    text_so2= so2.get()
    if text_so1.isdigit() == 'True' and text_so2.isdigit():
        tong = text_so1 + text_so2
        lb_kq['text']= f'{tong}'
    else:
        lb_kq['text'] = 'Error'
    
lb_so1 = tk.Label(master=window, text = 'Số 1 ')
lb_so1.place(x=10,y =20)
so1 = tk.Entry(master=window)
so1.place(x=50,y=20)
lb_so2 = tk.Label(master=window, text = 'Số 2 ')
lb_so2.place(x=10,y =60)
so2 = tk.Entry(master=window)
so2.place(x=50,y=60)
cal = tk.Button(master=window, text = 'Cal',command = kq)
cal.place ( x= 150, y = 100)
lb_kq = tk.Label(master=window, text = '')
lb_kq.place(x=50,y =100)
window.mainloop()
```

```python
from tkinter import *
top = Tk()
top.geometry("450x300")
def clear_error1(event):
    lb_error1['text']=''
def clear_error2(event):
    lb_error2['text']=''    
def changeText():
    valid_number1 = False
    valid_number2 = False
    try:
        number1 = float(txt_number1.get())
        valid_number1 = True
    except:
        lb_error1['text']='dữ liệu nhập vào chưa đúng'
    try:
        number2 = float(txt_number2.get())
        valid_number2 = True
    except:
        lb_error2['text']='dữ liệu nhập vào chưa đúng'
    if valid_number1 and valid_number2:
        lb_result['text']= number1 + number2
        lb_result['fg']= 'green'
        lb_error1['text']=''
        lb_error2['text']=''
    else:
        lb_result['text']= 'Lỗi dữ liệu đầu vào'
        lb_result['fg']= 'red'
myfont = ('Helvetica','16')
font_error = ('Helvetica','8', 'italic')
fontbutton = ('Helvetica','16', 'bold')
lb_number1 = Label(top,text = 'Số thứ nhất', font=myfont)
lb_number1.place(x=10,y = 20)
txt_number1= Entry (top, font = myfont, justify = LEFT,state=NORMAL)
txt_number1.place(x=130, y = 20)
txt_number1.bind('<FocusIn>',clear_error1)
lb_error1 = Label(top,text = 'Dữ liệu nhập vào chưa đúng', font=font_error,fg='red')
lb_error1.place(x=130, y = 50)

lb_number2 = Label(top,text = 'Số thứ hai', font=myfont)
lb_number2.place(x=10,y = 20)
txt_number2= Entry (top, font = myfont, justify = LEFT,state=NORMAL)
txt_number2.place(x=130, y = 80)
txt_number2.bind('<FocusIn>',clear_error2)
lb_error2 = Label(top,text = 'Dữ liệu nhập vào chưa đúng', font=font_error,fg='red')
lb_error2.place(x=130, y = 110)

lb_result = Label(top, text = 'Kết quả phép tính', font = myfont,fg = 'green',bg = '#cce7e8', width = 19, padx = 5)
lb_result.place(x=130,y=200)
button = Button(top, text = 'Tính tổng', command = changeText, font = fontbutton, fg ='blue',width = 18, padx = 3)
button.place (x=130,y=140)
top.mainloop()
```
