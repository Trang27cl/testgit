```python
#Nhập vào một chuỗi gồm các số tự nhiên cách nhau một khoảng trắng.
#Tính tổng, tính trung bình cộng các số. 
#Tìm số âm lớn nhất và số dương bé nhất trong chuỗi vừa nhập.
chuoi = input('Nhap chuoi ')
tap=chuoi.split()
tong=0
chuoiam=''
chuoiduong=''
for i in tap: 
    tong+=float(i)
    if float(i)<0:
        chuoiam= chuoiam + i + ' '
        tapmax=chuoiam.split()
        giatrimax=float(tapmax[0])
        for j in tapmax:
            if float(j) > giatrimax:
                giatrimax=float(j)
    if float(i)>0:
        chuoiduong= chuoiduong + i + ' '
        tapmin=chuoiduong.split()
        giatrimin=float(tapmin[0])
        for k in tapmin:
            if float(k) > giatrimax:
                giatrimin=float(k)       
TB= tong / (len(tap))
print('Tổng chuỗi là ',tong)
print('Trung bình cộng chuỗi là ',TB)
if chuoiam=='':
    print('Không có số âm')
else:
    print('Số âm lớn nhất ', giatrimax)
if chuoiduong=='':
    print('Không có số dương')
else:
    print('Số dương bé nhất ',giatrimin)
```
