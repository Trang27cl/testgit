```python
def ucln(a,b):
    '''Tìm ước chung lớn nhất của 2 số
    Parameters:
    a,b (int): Nhập 2 số
    Returns:
    a(int): ước chung 2 số'''
    while a!=b:
        if a>b:
            a=a-b
        else:
            b=b-a
    return a
def list_ucln(a,b,c):
    '''Tìm ước chung lớn nhất của 3 số
    Parameters:
    a,b,c (int): Nhập 2 số
    Returns:
    a(int): ước chung 2 số'''
    ab=ucln(a,b)
    ac=ucln(a,c)
    bc=ucln(b,c)
    return ab,ac,bc
result=list_ucln(12,6,22)
print(result)
```

```python
# Nhập vào số N, tìm tất cả các số nguyên tố nhỏ hơn N
def nguyento(n):
    '''Kiểm tra là số nguyên tố hay không
    Parameters:
    n (int): Nhập số tự nhiên n
    Returns:
    True/Flase: N là số tự nhiên là đúng/sai'''
    for i in range(2,int(n/2)+1,1):
        if n%i==0:
            return False # khi gặp lệnh return tương đương vs kết thúc, không cần lệnh break vòng lặp
    return True
n=int(input('Enter the number: '))
for i in range(2,n,1):
    if nguyento(i):
        print (i,end=' ')
```

```python
#BTVN2
import re
def scln(chuoi):
    '''Tìm số chẵn lớn nhất từ chuỗi
    Parameters:
    chuoi (str): Nhập chuỗi các kí tự bất kỳ gồm chữ cái và chữ số cách nhau bằng dấu cách
    Returns:
    max: số chẵn lớn nhất trong chuối hoặc không có số chẵn'''
    chuoiso=re.findall( r'\d+' , chuoi)
    chuoisochan=''
    for i in chuoiso:
        if int(i)%2==0:
            chuoisochan=chuoisochan+i+' '
    chuoisochan=chuoisochan.strip()
    if chuoisochan!='':
        chuoisochan=chuoisochan.split()
        max=int(chuoisochan[0])
        for j in chuoisochan:
            if int(j)>max:
                max=int(j)
        return(max)
    else:
        return('Không có số chẵn')
def slnn(chuoi):
    '''Tìm số lẻ nhỏ nhất từ chuỗi
    Parameters:
    chuoi (str): Nhập chuỗi các kí tự bất kỳ gồm chữ cái và chữ số cách nhau bằng dấu cách
    Returns:
    min: số chẵn lớn nhất trong chuối hoặc không có số chẵn'''
    chuoiso=re.findall( r'\d+' , chuoi)
    chuoisole=''
    for i in chuoiso:
        if int(i)%2==1:
            chuoisole=chuoisole+i+' '
    chuoisole=chuoisole.strip()
    if chuoisole!='':
        chuoisole=chuoisole.split()
        min=int(chuoisole[0])
        for j in chuoisole:
            if int(j)<min:
                min=int(j)
        return(min)
    else:
        return('Không có số lẻ')
b=scln('abd 34 tr13 85 sh 342 21 def')
a=slnn('abd 34 tr13 85 sh 342 21 def')
print(a,b)
```

```python
#BTVN1:
# Vẽ lịch:
import turtle
import calendar
import datetime
t=turtle.Turtle()
t.shape('turtle')
t.hideturtle()
t.color('black')
mm=int(input('Nhập tháng '))
yyyy=int(input('Nhập năm '))
#hàm di chuyển
def dichuyen(x,y):
    t.penup()
    t.goto(x,y)
    t.setheading(0)
    t.pendown()
# Vẽ tháng
month=['January','February','March','April','May','June','July','August','September','October','November','December']
if mm<0 and mm>13:
    print('Nhập tháng sai')
else:
    dichuyen(200,150)
    t.write(month[mm-1],align='right', font=('Arial', 80, 'normal'))
# Vẽ năm:
dichuyen(200,70)
t.color('red')
t.write(yyyy,align='right', font=('Arial', 60, 'italic'))
# Vẽ các ngày trong năm:
day_headings = ['S','M','T','W','T','F','S']
t.penup()
a=-100
b=0
dichuyen(a,b)
for i in range(7):
    t.pencolor("red") if i == 0 else t.pencolor("black")
    t.write(day_headings[i],align='right', font=('Arial', 20, 'normal'))
    a+=50
    dichuyen(a,b)
dichuyen(-120,-5)
t.pencolor("black")
t.forward(320)
#Vẽ lịch:
date1 = datetime.datetime(yyyy, mm, 1)
if mm <12:
    date2 = datetime.datetime(yyyy, mm+1, 1)
else:
    date2 = datetime.datetime(yyyy+1, 1, 1)
c=-100
d=-40
f=0
g=[1,2,3,4,5,6,0]
for i in range((date2-date1).days):
    date = datetime.datetime(yyyy,mm,i+1)
    e=date.weekday()
    z=g[e]
    dichuyen(c + z*50, d )
    t.pencolor("red") if z == 0 else t.pencolor("black")
    if e==5:
        d=d-40
    f+=1
    t.write(f, align='right', font=('Arial', 20, 'normal'))
t=turtle.done()
```
