```python
from tkinter import *
from tkinter import messagebox
root = Tk()
root.title('Combobox Demo')
root.geometry('400x200')

def selection():
    messagebox.showinfo ('Python course', pythonVariable.get())
    messagebox.showinfo ('Android course', javaVariable.get())
    messagebox.showinfo ('Javascript course', androidVariable.get())

pythonVariable=IntVar()
pythonVariable.set(1)
#Giá trị mặc định là Python course
javaVariable=IntVar()
androidVariable=IntVar()
checkButton1 = Checkbutton(root, text = 'Python (online course)', onvalue =1 , offvalue =0, variable = pythonVariable, command = selection)
checkButton1.place(x=10,y =20)
checkButton2 = Checkbutton(root, text = 'Android (offline course)', onvalue =1 , offvalue =0, variable = javaVariable, command = selection)
checkButton2.place(x=10,y =50)
checkButton3 = Checkbutton(root, text = 'JavaScript (project)', onvalue =1 , offvalue =0, variable = androidVariable, command = selection)
checkButton3.place(x=10,y =80)
root.mainloop()
```
