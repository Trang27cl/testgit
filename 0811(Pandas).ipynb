```python
#Thí sinh nào nhiều tuổi nhất, ít tuổi nhất
import datetime
data_stu['Age']=datetime.datetime.today().year-data_stu['Ngày sinh'].dt.year
data_stu.head()
#data_stu['Age']=datetime.datetime.today().year-pd.to_datetime(data_stu['Ngày sinh']).dt.year
```

```python
#Chia phổ điểm từ 0-10 thành 10 đoạn (0-1,1-2,2-3...9-10)
#Tính số luoengj sinh viên có điểm thi từng môn trong từng phổ điểm trên
spec = {(i,i+1):0 for i in range(10)}
#lấy biên bên trái không lấy biên bên phải
for i in data_stu['Toán'].values:
    if i<1:
        spec[(0,1)]+=1
    elif i<2:
        spec[(1,2)]+=1
    elif i<3:
        spec[(2,3)]+=1
    elif i<4:
        spec[(3,4)]+=1
    elif i<5:
        spec[(4,5)]+=1
    elif i<6:
        spec[(5,6)]+=1
    elif i<7:
        spec[(6,7)]+=1
    elif i<8:
        spec[(7,8)]+=1
    elif i<9:
        spec[(8,9)]+=1
    else:
        spec[(9,10)]+=1
spec
```

```python
data_stu['Đtb']=round((2*(data_stu['Toán']+data_stu['Văn']+data_stu['Tiếng Anh'])+(data_stu['Sử ']+data_stu['Địa']+data_stu['Lý']+data_stu['Hoá']+ data_stu['Sinh']))/11)
data_stu
```

```python
#Sinh viên đỗ trượt, đại học từng khối
data_stu['A'] = data_stu['Toán']+data_stu['Lý']+ data_stu['Hoá']
data_stu['A1'] = data_stu['Toán']+data_stu['Lý']+ data_stu['Tiếng Anh']
data_stu['B'] = data_stu['Toán']+data_stu['Sinh']+ data_stu['Hoá']
data_stu['C'] = data_stu['Văn']+data_stu['Sử ']+data_stu['Địa']
data_stu['D']= data_stu['Toán']+data_stu['Văn']+(data_stu['Tiếng Anh'])*2
data_stu
```

```python
data_stu['_A_']=data_stu['A']>=23.5
data_stu['_A1_']=data_stu['A1']>=22
data_stu['_B_']=data_stu['B']>=21.5
data_stu['_C_']=data_stu['C']>=22.5
data_stu['_D_']=data_stu['D']>=27.5
data_stu
```

```python
#so sánh điểm trung bình đại học
data_stu[data_stu.columns[13:18]].mean()
```

```python
#Xếp loại học sinh
import numpy as np
xuatsac = (data_stu['Đtb']>=9.0)&(data_stu['Toán']>=8)&(data_stu['Lý']>=8)&(data_stu['Hoá']>=8)&(data_stu['Sinh']>=8)&(data_stu['Văn']>=8)&(data_stu['Sử ']>=8)&(data_stu['Địa']>=8)&(data_stu['Tiếng Anh']>=8)
gioi = (data_stu['Đtb']>=8.0) &(data_stu['Toán']>=6.5)&(data_stu['Lý']>=6.5)&(data_stu['Hoá']>=6.5)&(data_stu['Sinh']>=6.5)&(data_stu['Văn']>=6.5)&(data_stu['Sử ']>=6.5)&(data_stu['Địa']>=6.5)&(data_stu['Tiếng Anh']>=6.5)
trungbinh = (data_stu['Đtb']>=6.0) &(data_stu['Toán']>=4.5)&(data_stu['Lý']>=4.5)&(data_stu['Hoá']>=4.5)&(data_stu['Sinh']>=4.5)&(data_stu['Văn']>=4.5)&(data_stu['Sử ']>=4.5)&(data_stu['Địa']>=4.5)&(data_stu['Tiếng Anh']>=4.5)
kem = data_stu['Đtb'] > 0.0

condition = [xuatsac, gioi, trungbinh, kem]
xl = ['Xuất sắc', 'Giỏi', 'Trung bình', 'Kém']
data_stu['Xếp loại'] = np.select(condition, xl)
data_stu
```
