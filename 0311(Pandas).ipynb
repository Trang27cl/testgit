```python
import random
import pandas as pd
numbers = [i for i in range(100)]
data = pd.Series([random.choice(numbers) for _ in range(100)])
data.values
```

```python
data
```

```python
min_value = data.iloc[0]
for k in data.values:
    if k < min_value: 
        min_value = k
print('Số nhỏ nhất ', min_value)
```

```python
max_value = data.iloc[0]
for j in data.values:
    if j > max_value:
        max_value = j
print('Số lớn nhất ',max_value)
```

```python
data.min()
```

```python
max_value = data.loc[0]
min_value = data.loc[0]
for i in list(data.index):
    if max_value < data.loc[i]:
        max_value = data.loc[i]
    if min_value > data.loc[i]:
        min_value = data.loc[i]
print(max_value,min_value)
```

```python
max_value = data.loc[0]
min_value = data.loc[0]
for index,val in list(data.items()):
    if max_value < val:
        max_value = val
    if min_value > val:
        min_value = val
print(max_value,min_value)
```

```python
#data.count hoặc data.len
max_value = data.loc[0]
min_value = data.loc[0]
for i in range(data.count()):
    if max_value < data.iloc[i]:
        max_value = data.iloc[i]
    if min_value > data.iloc[i]:
        min_value = data.iloc[i]
print(max_value,min_value)
```

```python
data.mean()
```

```python
above_mean = 0
below_mean = 0
mean_val = data.mean()
for i in range(data.count()):
    if data.iloc[i] > mean_val:
        above_mean +=1
    else:
        below_mean +=1
print(above_mean,below_mean)
```

```python
sum_value = 0
for x in data:
    sum_value = sum_value + x
print(sum_value)
average_value = sum_value / 100
print(average_value)
slgtn=0
slgtl=0
for a in data:
    if a < average_value:
        slgtn=slgtn + 1
print('Số lượng nhỏ hơn giá trị trung bình',slgtn)
for b in data:
    if b > average_value:
        slgtl=slgtl + 1
print('Số lượng lớn hơn giá trị trung bình',slgtl)
```

```python
slsc=0
for c in data:
    if c%2 == 0:
        slsc=slsc +1
print('Số lượng số chẵn',slsc)
```

```python
slsl=0
for d in data:
    if d %2 !=0:
        slsl = slsl +1
print('Số lượng số lẻ',slsl)
```

```python
slsa = 0
for e in data:
    if e <0:
        slsa = slsa +1
print('Số lượng số âm',slsa)
```

```python
slsd = 0
for f in data:
    if f > 0:
        slsd = slsd +1
print('Số lượng số dương',slsd)
```

```python
#Số lần xuất hiện các số trong Series
c=0
slxh={}
for a in range (len(data)):
    for b in data:
        if b == data.iloc[a]:
            c = c+1
    slxh [data.iloc[a]]=c
    c=0
print(slxh)
```

```python
# Tìm tần số xuất hiện của các phần tử trong Series
dicts = {i:0 for i in range(100)}

for i in range(data.count()):
    if data.iloc[i] in dicts.keys():
        dicts[data.iloc[i]]+=1
dicts
result = sorted(dicts.items(), key=lambda x: x[1], reverse = True)
#x[1] chỉ cột giá trị, x từng phần tử trong items; x[0] chỉ cột keys, reverse = True theo chiều giảm dần
result = dict(result)
result
```
