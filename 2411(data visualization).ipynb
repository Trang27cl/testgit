```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Chia từ 0 đến 10 thành 100 đoạn bằng nhau
x = np.linspace (0,10,100)
x
```

```python
#Biểu đồ đường
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x), label ='sin(x)')
plt.plot(x,np.cos(x), label ='cos(x)')
plt.title ('Hàm sin(x), cos (x)...', fontdict={'fontsize':12,'color':'blue'})
plt.xlabel('x values')
plt.ylabel('sin(x), cos(x) values')
plt.legend()
plt.show()
```

```python
#Thay đổi cài đặt Axes,Axis
axes=plt.gca()
axes.set_xlabel('Giá trị đầu vào',color='red',fontsize=14)
axes.set_ylabel('Giá trị đầu ra',color='green',fontsize=14)
axes.tick_params(axis='x', colors='blue',size=12)
axes.tick_params(axis='y', colors='black',size=12)
plt.show()
```

```python
#Adjusting lines color and style
plt.plot(x,np.sin(x - 0),color='blue') #specify color by name
plt.plot(x,np.sin(x - 1),color='g') #short color by code (rgbcmyk)
plt.plot(x,np.sin(x - 2),color='0.75') #Grayscale between 0 and 1
plt.plot(x,np.sin(x - 3),color='#a150ec') #Hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x - 4),color=(1.0,0.2,0.3)) #RGB tuple, values 0 and 1
plt.plot(x,np.sin(x - 5),color='chartreuse') # all HTML color names supported
```

```python
plt.plot(x,x+0, linestyle='solid')
plt.plot(x,x+1, linestyle='dashed')
plt.plot(x,x+2, linestyle='dashdot')
plt.plot(x,x+3, linestyle='dotted')
```

```python
plt.plot(x,x+4, linestyle='-') #solid
plt.plot(x,x+5, linestyle='--') #dashed
plt.plot(x,x+6, linestyle='-.') #dashdot
plt.plot(x,x+7, linestyle=':') #dotted
```

```python
#Axes Limits
plt.plot(x,np.sin(x),color='r',linestyle='--')
plt.show()
```

```python
#Tăng độ dài x,y
plt.plot(x,np.sin(x),color='r',linestyle='--')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
#Đối xứng qua truc X 
#reverse order
plt.plot(x,np.sin(x),color='r',linestyle='--')
plt.xlim(-1,11)
plt.ylim(1.5,-1.5)
```

```python
rng = np.random.RandomState(0)
x= rng.randn(100)
y= rng.randn(100)
```

```python
#Độ đậm nhạt
colors = rng.rand(100)
colors
```

```python
#Độ to nhỏ giá trị dựa vào size
sizes=1000* rng.rand(100)
plt.scatter(x,y,c=colors, marker='o',s=sizes,alpha = 0.6, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#Histograms: thấy được độ phân bố, tần suất phân bố dữ liệu trong 1 khoảng nào đó
data = np.random.randn(1000)
plt.hist(data)
plt.show()
#Randn: xugn quanh 0, có cả âm, rand: không có âm
```

```python
plt.hist(data, bins=30,alpha = 0.5, histtype='stepfilled',color='steelblue',edgecolor='red')
#edgecolor : màu viền,alpha: độ đậm nhạt
#bins: số khoảng, số steps
```

```python
#kwargs: tham số truyền vào
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(-3,2,1000)
kwargs=dict(histtype='stepfilled',alpha = 0.3, bins=40)
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
plt.show()
```

```python
#Pie chart thường dùng để thấy phần trăm, tỉ lệ (thường category ít hơn 6)
labels = ['Gạo','Gas','Thịt', 'Trứng','Sữa']
sizes = [443,292,231,110,20]
plt.pie(sizes, labels = labels)
plt.title('Sản lượng xuất khẩu năm 2021', fontsize = 14)
plt.show()
```

```python
under_five = students['Toán'][students['Toán']<=5].count()
```

```python
five_to_eight=students['Toán'][(students['Toán']<=8)&(students['Toán']>5)].count()
```

```python
greater_eight = students['Toán'][students['Toán']>8].count()
```

```python
scores=[under_five,five_to_eight,greater_eight]
labels = ['Kém','Khá','Giỏi']
plt.pie(scores,labels=labels)
plt.title('Phân loại điểm toán')
plt.show()
```

```python
#Customize pie chart
#Cài đặt màu sắc cho các thành phần:
plt.pie(scores,labels=labels, colors=['r','g','y'])
plt.show()
```

```python
#Dùng để gán nhãn % cho từng thành phần:
plt.pie (scores, labels = labels, autopct='%1.2f%%')
plt.show()
```

```python
#Tỷ lệ giữa tâm của mỗi phần và nhãn (autopct !=None):
plt.pie(scores, labels = labels, autopct='%1.2f%%', pctdistance = 0.4)
plt.show()
```

```python
#Vẽ bóng mờ cho các thành phần:
plt.pie(scores, labels = labels, shadow = True)
plt.show()
```

```python
#bán kính của hình tròn
plt.pie(scores, labels = labels, radius = 1.0)
plt.show()
```
