```python
class Circle:
    #Hàm khởi tạo của lớp: Khởi tạo các thuộc tính của lớp
    def __init__(self,r):
        self.r = r
    #hàm thành viên của lớp
    def area(self):
        return 3.14*self.r*self.r
    def info(self):
        print(f'r = {self.r}')
    #Hoàn thành định nghĩ 1 lớp
```

```python
circle = Circle(10)
dt = circle.area()
print(dt)
```

```python
#Tạo ra 1 list đường tròn có bán kính bất kì
#Tìm đường tròn có bán kính lớn nhất
#Sắp xếp list đường tròn theo bán kính
import random
import math
list_circle =[]
max_dt = 0
for i in range(5):
    r = random.randint(1,100)
    circle = Circle(r)
    dt=circle.area()
    list_circle.append(dt)
for i in list_circle:
    if i >= max_dt:
        max_dt=i
print(max_dt)
list_circle=sorted(list_circle)
print(list_circle)
```

```python
# Định nghĩa lớp sinh viên Student gồm các thông tin: mã sinh viên, họ tên, ngày sinh, quê quán.
# Phương thức info để in thông tin của sinh viên ra màn hình.
# Định nghĩa lớp danh sách sinh viên ListStudent thuộc tính là một list Student vừa định nghĩa ở trên.
# Các phương thức thêm sinh viên mới, cập nhật thông tin sinh viên theo mã sinh viên, xóa sinh viên trong list, tìm kiếm thông tin theo mã, in toàn bộ danh sách sinh viên ra màn hình.
# Viết 1 phần mềm quản lý danh sách sinh viên có các chức năng sau:
# 1. Xem danh mục sinh viên
# 2. Thêm sinh viên vào danh sách
# 3. Chỉnh sửa thông tin sinh viên
# 4. Xóa thông tin sinh viên trong danh sách
# 5. Tìm kiếm thông tin sinh viên
# 6. Sắp xếp theo các tiêu chí: Tên, mã ...
class Student:
    def __init__(self,code, name, dateofbirth,residence):
        self.code = code
        self.name = name
        self.dateofbirth = dateofbirth
        self.residence = residence
    def info(self):
        row = self.code.ljust(20, ' ') + self.name.ljust(25, ' ') + self.dateofbirth.ljust(25, ' ') + self.residence
        print(row)
```

```python
# Định nghĩa lớp sinh viên Student gồm các thông tin: mã sinh viên, họ tên, ngày sinh, quê quán.
# Phương thức info để in thông tin của sinh viên ra màn hình.
# Định nghĩa lớp danh sách sinh viên ListStudent thuộc tính là một list Student vừa định nghĩa ở trên.
# Các phương thức thêm sinh viên mới, cập nhật thông tin sinh viên theo mã sinh viên, xóa sinh viên trong list, tìm kiếm thông tin theo mã, in toàn bộ danh sách sinh viên ra màn hình.
# Viết 1 phần mềm quản lý danh sách sinh viên có các chức năng sau:
# 1. Xem danh mục sinh viên
# 2. Thêm sinh viên vào danh sách
# 3. Chỉnh sửa thông tin sinh viên
# 4. Xóa thông tin sinh viên trong danh sách
# 5. Tìm kiếm thông tin sinh viên
# 6. Sắp xếp theo các tiêu chí: Tên, mã ...
class Student:
    def __init__(self,code, name, dateofbirth,residence):
        self.code = code
        self.name = name
        self.dateofbirth = dateofbirth
        self.residence = residence
    def info(self):
        row = self.code.ljust(20, ' ') + self.name.ljust(25, ' ') + self.dateofbirth.ljust(25, ' ') + self.residence
        print(row)
```

```python
class ListStudent:
    def __init__(self,students):
        self.L = students
    def load(self):
        #Load từ file ra
        stud1 = Student('20225115', 'Lê Chí Bách', '22/06/2000','Hà Nội')
        stud2 = Student('20225116', 'Võ Thị Phượng', '12/07/1995','Hà Nam')
        stud3 = Student('20225117', 'Đặng Quang Sáng', '05/12/1999','Bắc Ninh')
        self.L = [stud1, stud2, stud3]
    def info(self):
        header = 'Mã sinh viên'.ljust(20, ' ') + 'Họ tên'.ljust(25, ' ') + 'Ngày sinh'.ljust(25, ' ') + 'Quê quán'
        print(header)
        print(''.ljust(80, '='))
        for stud in self.L:
            stud.info()
    def add(self,st):
        self.L.append(st)
    def delete(self,deleting_code):
        succeeded = False
        for st in self.L:
            if st.code == deleting_code:
                confirm = input(f'Bạn chắc chắn muốn xóa sinh viên có mã {deleting_code}?')
                if confirm == 'y' or confirm == 'Y':
                    self.L.remove(st)
                    succeeded = True
                    break
        return succeeded
    def handle(self,handle_code):
        for sv in self.L:
            if sv.code == handle_code:
                confirm = input(f'Bạn chắc chắn muốn chỉnh sửa sinh viên có mã {handle_code}?')
                if confirm == 'y' or confirm == 'Y':
                    while 1:
                        print('Nhập thông tin bạn muốn chỉnh sửa ')
                        print('1. Mã sinh viên')
                        print('2. Họ tên sinh viên')
                        print('3. Ngày sinh sinh viên')
                        print('4. Quê quán sinh viên')
                        print('5. Kết thúc chỉnh sửa')
                        choice = input(' Nhập lựa chọn: ')
                        if choice =='1':
                            code = input('Nhập mã sinh viên: ')
                            sv.code = code
                        elif choice =='2':
                            name = input('Nhập họ tên sinh viên: ')
                            sv.name = name
                        elif choice =='3':
                            dateofbirth = input('Nhập ngày sinh sinh viên: ')
                            sv.dateofbirth = dateofbirth
                        elif choice =='4':
                            residence = input('Nhập quê quán sinh viên: ')
                            sv.residence = residence
                        elif choice =='5':
                            confirm = input('Bạn muốn kết thúc chỉnh sửa (y/n)?')
                            if confirm =='y'or confirm =='Y':
                                break
                    print('Kết thúc chỉnh sửa')
    def search(self, search_code):
        succeeded = False
        for sv in self.L:
            if sv.code == search_code:
                row = sv.code.ljust(20, ' ') + sv.name.ljust(25, ' ') + sv.dateofbirth.ljust(25, ' ') + sv.residence
                print(row)
                succeeded = True
        return succeeded
    def sort_name(self):
        self.L = sorted(self.L, key = lambda x: x.name)
        header = 'Mã sinh viên'.ljust(20, ' ') + 'Họ tên'.ljust(25, ' ') + 'Ngày sinh'.ljust(25, ' ') + 'Quê quán'
        print(header)
        print(''.ljust(80, '='))
        for stud in self.L:
            stud.info()
    def sort_code(self):
        self.L = sorted(self.L, key = lambda x: x.code)
        header = 'Mã sinh viên'.ljust(20, ' ') + 'Họ tên'.ljust(25, ' ') + 'Ngày sinh'.ljust(25, ' ') + 'Quê quán'
        print(header)
        print(''.ljust(80, '='))
        for stud in self.L:
            stud.info()
```

```python
class Program:
    def __init__(self):
        self.dssv = ListStudent(None)
        self.dssv.load()
    def run(self):
        while 1:
            print('========== QUẢN LÝ SINH VIÊN ==========')
            print('======== 1. Hiển thị danh sách ========')
            print('========== 2. Thêm sinh viên ==========')
            print('======== 3. Chỉnh sửa thông tin =======')
            print('====== 4. Xóa thông tin sinh viên =====')
            print('======== 5. Tìm kiếm thông tin ========')
            print('========= 6. Sắp xếp theo tên =========')
            print('===== 7. Sắp xếp theo mã sinh viên ====')
            print('========= 8. Đóng chương trình ========')
            print('=======================================')
            choice = input(' Nhập lựa chọn: ')
            if choice =='1':
                self.dssv.info()
                input('Nhấn enter để tiếp tục ...')
            elif choice =='2':
                code = input('Nhập mã sinh viên: ')
                name = input('Nhập họ tên sinh viên: ')
                date = input('Nhập ngày sinh sinh viên: ')
                residence = input('Nhập quê quán sinh viên: ')
                new_stud = Student(code, name, date, residence)
                self.dssv.add(new_stud)
                print('Thêm sinh viên mới thành công!')
                input('Nhấn enter để tiếp tục ...')
            elif choice =='3':
                code = input('Nhập mã sinh viên muốn chỉnh sửa: ')
                self.dssv.handle(code)
            elif choice =='4':
                deleting = input('Nhập mã sinh viên cần xóa: ')
                succeeded = self.dssv.delete(deleting)
                if succeeded:
                    print('Xóa thành công...')
                else:
                    print(f'Xóa thất bại. Có thể mã sinh viên {deleting} không tồn tại')
                input('Nhấn enter để tiếp tục ...')
            elif choice =='5':
                searching = input('Nhập mã sinh viên cần tìm: ')
                sc = self.dssv.search(searching)
                if sc == False:
                    print(f'Tìm kiếm thất bại. Có thể mã sinh viên {searching} không tồn tại')
                input('Nhấn enter để tiếp tục ...')
            elif choice =='6':
                self.dssv.sort_name()
                input('Nhấn enter để tiếp tục ...')
            elif choice =='7':
                self.dssv.sort_code()
                input('Nhấn enter để tiếp tục ...')
            elif choice =='8':
                confirm = input('Bạn muốn đóng chương trình (y/n)?')
                if confirm =='y'or confirm =='Y':
                    break
        print('Chương trình kết thúc')
```

```python
main = Program()
main.run()
```
