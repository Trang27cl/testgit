```python
# thay thế dữ liệu khuyết thiếu bởi giá trị nội suy theo cột
df2 = df.interpolate(axis=1)
```

```python
#Thay thế giá trị khuyết thiếu bằng giá trị 0
df3=df.fillna(0)
df3
```

```python
# vẽ biểu đồ box plot cho MonthlyIncome
sns.boxplot(df2["MonthlyIncome"])
```

```python
# tính giá trị Q1 và Q3
Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
```

```python
# tính IQR 
IQR = Q3-Q1
```

```python
# lọc dữ liệu ngoại lai
df3 = df2[~((df2 < (Q1 - 1.5 * IQR)) | (df2 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df3.boxplot()
```

```python
sns.boxplot(df3["MonthlyIncome"])
```

```python
# chuẩn hóa với minmax scaling
scaler = MinMaxScaler()

mms = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
# chuẩn hóa với robust scaling
scaler = RobustScaler()

rbs = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
# chuẩn hóa với standard scaling
scaler = StandardScaler()

sc = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
#Chia dữ liệu các cột thành  4,5,6 nhóm có số lượng phần tử bằng nhau và đếm số lượng phần tử ở mỗi nhóm, lấy ra khoảng giữ liệu của mỗi nhóm và đặt tên cho các nhóm
cats = pd.qcut(df3['MonthlyIncome'], 4,labels=['Poor','normal','Good','Excellent'])
cats
```

```python
# số lượng phần từ ở mỗi phần
pd.value_counts(cats)
```

```python
#Chia dữ liệu ở các cột age  thành 5 nhóm theo các khoảng: 0, 30, 40, 50, 80, 150 và đếm số lượng phần tử ở mỗi nhóm, đặt tên cho các nhóm
cut_month= pd.cut(df3['age'], [0,30,40,50,80,150],labels= ['Youth', 'YoungAdult', 'MiddleAged', 'Senior','elder'])
cut_month
```

```python
# số lượng phần từ ở mỗi phần
pd.value_counts(cut_month)
```
