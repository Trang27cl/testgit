```python
from tkinter import *
from tkinter import messagebox
#Text dùng để nhập text - nó khác vs entry ở chỗ là text cho phép nhập text nhiều dòng kahsc nhau
# còn entry chỉ nhập được 1 dòng duy nhất 
#label frame
def callback():
    #Text chứa nhiều dòng: từ 1 cho đến hết
    messagebox.showinfo('Info', text.get(1.0,'end'))
top = Tk()
lb_frame1 = LabelFrame(top, text = 'Text widget instance')
text = Text (lb_frame1)
text.pack(padx = 5, pady = 5, fill = BOTH)
lb_frame1.pack(fill=BOTH)

lb_frame2 = LabelFrame(top, text = 'Entry widget instance')
single_input = Entry (lb_frame2)
single_input.pack(pady = 20)
lb_frame2.pack(fill=BOTH)

button = Button(top, text = 'Show text', command=callback)
button.pack()
top.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

root = Tk()
root.title('Tính toán 2 số')
root.geometry('500x300')

lbfont = ('Helvetica','16')

def tinhtoan(event):
    gt1= float(so1.get())
    gt2= float(so2.get())
    if var_thaotac.get()=='+':
        kq=gt1+gt2
        lb_ketqua['text']= f'{kq}'
    elif var_thaotac.get()=='-':
        kq=gt1-gt2
        lb_ketqua['text']= f'{kq}'
    elif var_thaotac.get()=='x':
        kq=gt1*gt2
        lb_ketqua['text']= f'{kq}'
    else:
        if gt2!=0:
            kq=round(gt1/gt2,2)
            lb_ketqua['text']= f'{kq}'
        else:
            lb_ketqua['text']= 'Error'

lb_so1 = Label(root, text = 'So1', fg = 'red',  font=lbfont)
lb_so1.place(x=10,y =20)

so1 = Entry (root, font=lbfont, justify = RIGHT)
so1.place(x= 110, y =20)

lb_so2 = Label(root, text = 'Số 2', fg = 'red',  font=lbfont)
lb_so2.place(x=10,y =70)

so2 = Entry (root, font=lbfont, justify = RIGHT)
so2.place(x= 110, y =70)

lb_thaotac = Label(root, text = 'Thao tác', fg = 'red',  font=lbfont)
lb_thaotac.place(x=10,y =120)

values = ['+','-','x','/']
var_thaotac=StringVar()
combobox_thaotac = ttk.Combobox(root, values = values, textvariable = var_thaotac, font=lbfont, width = 18)
combobox_thaotac.place(x=112,y =120)
combobox_thaotac.set(values[0])

button = Button(root, text = 'Tính', fg = 'red',  font=lbfont)
button.bind('<Button>',tinhtoan)
button.place(x = 110, y= 170)

lb_ketqua = Label(root, text = '', fg = 'red',  font=lbfont)
lb_ketqua.place(x=250,y =170)
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
root = Tk()
root.title('Tính thuế thu nhập')
root.geometry('600x300')

lbfont = ('Helvetica','16')

def nhaythue(event):
    try:
        tongthunhap=float(thunhap.get())
        songuoiphuthuoc=int(nguoiphuthuoc.get())
        thunhaptinhthue=tongthunhap-songuoiphuthuoc*4400000-11000000
        if thunhaptinhthue>0:
            if thunhaptinhthue <= 5000000:
                combobox_thaotac.set(values[1])
            elif thunhaptinhthue <= 10000000:
                combobox_thaotac.set(values[2])
            elif thunhaptinhthue <= 18000000:
                combobox_thaotac.set(values[3])
            elif thunhaptinhthue <= 32000000:
                combobox_thaotac.set(values[4])
            elif thunhaptinhthue <= 52000000:
                combobox_thaotac.set(values[5])
            elif thunhaptinhthue <= 80000000:
                combobox_thaotac.set(values[6])
            else:
                combobox_thaotac.set(values[7])
        else:
            combobox_thaotac.set(values[0])
    except:
        combobox_thaotac.set(values[0])
def tinhthue(event):
    tongthunhap=float(thunhap.get())
    songuoiphuthuoc=int(nguoiphuthuoc.get())
    thunhaptinhthue=tongthunhap-songuoiphuthuoc*4400000-11000000
    if thunhaptinhthue>0:
        if thunhaptinhthue <= 5000000:
            ThueTNCN= thunhaptinhthue*0.05
        elif thunhaptinhthue <= 10000000:
            ThueTNCN= (5000000*0.05) + ((thunhaptinhthue-5000000)*0.1)
        elif thunhaptinhthue <= 18000000:
            ThueTNCN= (5000000*0.05) + (5000000*0.1) + ((thunhaptinhthue-10000000)*0.15)
        elif thunhaptinhthue <= 32000000:
            ThueTNCN= (5000000*0.05) + (5000000*0.1) + (8000000*0.15) + ((thunhaptinhthue-18000000)*0.2)
        elif thunhaptinhthue <= 52000000:
            ThueTNCN= (5000000*0.05) + (5000000*0.1) + (8000000*0.15) + (14000000*0.2) + ((thunhaptinhthue-32000000)*0.25)
        elif thunhaptinhthue <= 80000000:
            ThueTNCN= (5000000*0.05) + (5000000*0.1) + (8000000*0.15) + (14000000*0.2) + (20000000*0.25 )+ ((thunhaptinhthue-52000000)*0.3)
        else:
            ThueTNCN= (5000000*0.05) + (5000000*0.1) + (8000000*0.15) + (14000000*0.2)
        lb_ketqua['text']=f'{ThueTNCN}'
    else:
        lb_ketqua['text']='Không đóng thuế'
lb_thunhap = Label(root, text = 'Tổng thu nhập', fg = 'red',  font=lbfont)
lb_thunhap.place(x=10,y =20)

thunhap = Entry (root, font=lbfont, justify = RIGHT)
thunhap.place(x= 200, y =20)
thunhap.bind('<Key>',nhaythue)
#KEY HAY FOCUSOUT?
lb_nguoiphuthuoc = Label(root, text = 'Số người phụ thuộc', fg = 'red',  font=lbfont)
lb_nguoiphuthuoc.place(x=10,y =70)

nguoiphuthuoc = Entry (root, font=lbfont, justify = RIGHT)
nguoiphuthuoc.place(x= 200, y =70)
nguoiphuthuoc.bind('<Key>',nhaythue)

lb_thuesuat = Label(root, text = 'Thuế suất', fg = 'red',  font=lbfont)
lb_thuesuat.place(x=10,y =120)

values = ['0%','5%','10%','15%','20%','25%','30%','35%']
var_thaotac=StringVar()
combobox_thaotac = ttk.Combobox(root, values = values, textvariable = var_thaotac, font=lbfont, width = 18)
combobox_thaotac.place(x=202,y =120)
combobox_thaotac.set(values[0])


button = Button(root, text = 'Tính thuế', fg = 'red',  font=lbfont)
button.bind('<Button>',tinhthue)
button.place(x = 200, y= 170)

lb_ketqua = Label(root, text = '', fg = 'red',  font=lbfont)
lb_ketqua.place(x=300,y =170)
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

root = Tk()
root.geometry('350x250')
income_on_tax = 0

def tax_estimate():
    global income_on_tax
    tax_scale_val = tax_variable.get()[:-1]
    tax_paid = income_on_tax*float(tax_scale_val)/100
    button['text'] = f'Tính TNCN: {tax_paid}'
    
def tax_scale_selection(event):
    global income_on_tax
    try:
        total_income_val = float(total_income.get())
    except:
        lb_total_income_error['text']='*'
        return
    try:
        if total_child.get()=='':
            total_child_val = 0
        else:
            total_child_val = int(total_child.get())
    except:
        lb_total_income_error['text']='*'
        return
    income_on_tax = total_income_val - 11_000_000 - total_child_val*4_000_000
    
    if income_on_tax <= 5_000_000:
        income_on_tax_selection = 0
    elif income_on_tax <= 10_000_000:
        income_on_tax_selection = 1
    elif income_on_tax <= 18_000_000:
        income_on_tax_selection = 2
    elif income_on_tax <= 32_000_000:
        income_on_tax_selection = 3
    elif income_on_tax <= 52_000_000:
        income_on_tax_selection = 4
    elif income_on_tax <= 80_000_000:
        income_on_tax_selection = 5
    else:
        income_on_tax_selection = 6
    tax_variable.set(tax_scale[income_on_tax_selection])

lb_total_income = Label (root, text = 'Nhập tổng thu nhập')
lb_total_income.place(x=10, y= 20)

lb_total_income_error = Label (root, text = '',fg='red', font = ('Helvetica','12','bold'))
lb_total_income_error.place(x=290, y = 50)

total_income = Entry(root)
total_income.place(x=160, y= 20)
total_income.bind('<FocusOut>', tax_scale_selection)

lb_total_child = Label (root, text = 'Nhập số người phụ thuộc')
lb_total_child.place(x=10, y= 50)

total_child = Entry(root)
total_child.place(x=160, y= 50)
total_child.bind('<FocusOut>', tax_scale_selection)

tax_scale = [f'{i}%' for i in range(5,40,5)]
tax_variable = StringVar()

lb_tax_scale = Label (root, text = 'Thuế suất')
lb_tax_scale.place(x=90, y= 80)
c_tax_scale = ttk.Combobox(root, values = tax_scale, textvariable = tax_variable, state = 'readonly')
c_tax_scale.place(x=160, y= 80)

button = Button(root, text = 'Tính TNCN', command = tax_estimate)
button.place(x=10, y= 110, width = 292)

root.mainloop()
```
