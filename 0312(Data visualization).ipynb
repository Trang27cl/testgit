```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
#titanic dữ liệu thống kê về titanic
#iris dữ liệu về hoa
```

```python
#Load bộ dữ liệu tips có sẵn trong seaborn:
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Styling and Themes in Seaborn: set_style(). Some popular themes: darkgrid, whitegrid, dark,white,ticks,...
```

```python
plt.scatter(tips.total_bill,tips.tip)
sns.set_style('darkgrid')
plt.show()
```

```python
# Biểu đồ xu hướng cho thấy xu hướng dữ liệu, biểu diễn mô hình tuyến tính
#seaborn.lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue = 'smoker',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue = 'smoker',\
           data=tips, markers=['o','s'],palette ='Set1',legend = True)
#palette: bảng màu
plt.show()
```

```python
#Số col,row phụ thuộc vào số nhóm dữ liệu
sns.lmplot(x='total_bill',y='tip',col = 'smoker',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',row = 'smoker',data=tips)
plt.show()
```

```python
#Biểu đồ phân bố: sns.violinplot:
#Tổng hợp boxplot và độ phân bố dữ liệu
```

```python
sns.violinplot(y=tips['total_bill'],palette='coolwarm')
```

```python
#Sự phân bố total_bill theo ngày
sns.violinplot(x='day',y='total_bill',data=tips)
#Đường đậm ngày thứ 7, cn nhỉnh hơn t5,t6
```

```python
#Sự phân bố total_bill theo ngày và theo smoker
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='muted')
#hút thuốc dao động nhiều hơn
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='muted',split=True)
#Nó cắt một nửa thông tin này ghép với cái còn lại -> gọn hơn
```

```python
#Kiểm soát thứ tự các violinplot bằng cách truyền một thứ tự nhãn cụ thể:
sns.violinplot(x='time',y='total_bill',hue='smoker',data=tips,palette='muted',order =['Dinner','Lunch'])
```

```python
sns.violinplot(x='time',y='total_bill',hue='smoker',data=tips,palette='muted',order =['Lunch','Dinner'])
```

```python
#Biểu đồ tần số, không cần tính count
sns.countplot(x='sex',data=tips)
```

```python
female = tips[tips['sex']=='Female']['sex'].count()
male = tips[tips['sex']=='Male']['sex'].count()
plt.bar(['Female','Male'],[female,male])
plt.show()
```

```python
sns.countplot(x='sex',hue = 'smoker', data=tips)
```

```python
#Vẽ biểu đồ theo chiều ngang
sns.countplot(y='sex',hue = 'smoker', data=tips)
```

```python
#Biểu đồ heatmap: thể hiện tần số, độ phân bố, giá trị của từng điểm dữ liệu
uniform_data = np.random.rand(10,12) #tạo ma trận 10 hàng, 12 cột
uniform_data
```

```python
sns.heatmap(uniform_data)
```

```python
#Thay đổi giới hạn màu sắc
sns.heatmap(uniform_data,vmin = 0, vmax= 10)
```

```python
#biểu đồ boxplot
sns.boxplot(x=tips['total_bill'])
```

```python
#Vẽ phân nhóm theo ngày
sns.boxplot(x='day', y= 'total_bill',data=tips)
```

```python
#Vẽ phân nhóm theo nhiều nhóm
sns.boxplot(x='day', y= 'total_bill', hue = 'smoker', data=tips,palette='Set3')
```

```python
sns.boxplot(x='day', y= 'total_bill', hue = 'time', data=tips,linewidth=2.5)
#linewidth;: độ đậm nhạt của đường
```

```python
#Biểu đồ cặp: seaborn.pairplot, sử dụng khi dữ liệu ít do chạy lâu, rườm rà, dữ liệu lặp nhau
sns.pairplot(tips, hue = 'sex')
```
