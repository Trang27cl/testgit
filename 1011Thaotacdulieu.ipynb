```python
import numpy as np
students['ĐH khối A']=np.where((students['Toán']+students['Lý']+\
                                 students['Hoá'])>23.5,'Đỗ','Trượt')
students['ĐH khối A1']=np.where((students['Toán']+students['Lý']+\
                                 students['Tiếng Anh'])>22,'Đỗ','Trượt')
students['ĐH khối B']=np.where((students['Toán']+students['Hoá']+\
                                 students['Sinh'])>21.5,'Đỗ','Trượt')
students['ĐH khối C']=np.where((students['Văn']+students['Sử ']+\
                                 students['Địa'])>22.5,'Đỗ','Trượt')
students['ĐH khối D']=np.where((students['Toán']+students['Văn']+\
                                 2*(students['Tiếng Anh']))>27.5,'Đỗ','Trượt')

students
```

```python
#nhiều lựa chọn sử dụng np.select
xs = students['Toán']>=9
gioi = (students['Toán']>=8.0) &(students['Toán']<9)
kha = (students['Toán']>=7.0) &(students['Toán']<8)
tb = (students['Toán']>=5.0) &(students['Toán']<7)
conditions = [xs, gioi, kha, tb]
xl = ['Xuất sắc', 'Giỏi', 'Khá', 'Trung bình']
students['XL Toán'] = np.select(conditions, xl,default='Kém')
students
```
